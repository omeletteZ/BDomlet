```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/result.csv')
print(f"Всего записей: {len(df)}")
df.head()
```

 превращаем в список строк

```python
lines = df.apply(lambda row: ' '.join(str(v) for v in row.values), axis=1).tolist()
print(f"Всего строк: {len(lines)}")
print("Первые 3:")
for l in lines[:3]:
    print(l)
```

Задание 1 — перевести все слова в UPPERCASE через map

```python
upper_lines = list(map(lambda line: line.upper(), lines))
print("Первые 3 в uppercase:")
for l in upper_lines[:3]:
    print(l)
```

Задание 2 — оставить только строки где есть слова из white_list

```python
white_list = ['INSERT', 'UPDATE', 'DELETE']

def white_filter(line):
    words = line.split()
    matched = [w for w in words if w in white_list]
    return ' '.join(matched)

filtered_lines = list(map(white_filter, filter(lambda line: any(w in line for w in white_list), lines)))

print("Результат:")
for l in filtered_lines[:5]:
    print(l)
```

Задание 3 — посчитать сколько раз встречается каждое слово из white_list

```python
from collections import Counter
import functools

all_words = []
for line in lines:
    all_words += line.split()
white_words = list(filter(lambda w: w in white_list, all_words))

counts = Counter(white_words)
print("Количество каждого типа операции:")
print(counts)
```
